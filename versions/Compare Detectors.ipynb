```python
%config InlineBackend.figure_format = 'svg'

import matplotlib.pyplot as plt

from src.video import Video
```

```python
compare_frames = [20, 120, 220]
test_number = 4

save_name = f'test{test_number}_compare.png'

video1 = Video(f'results/one_class_YOLOV5s/test{test_number}.mp4')
video2 = Video(f'results/person_military_seperate_YOLOv5l/test{test_number}.mp4')

print(video1.num_frames, video2.num_frames)

frames1, frames2, = [], []

for video, frames in zip([video1, video2], [frames1, frames2]):
    for idx, frame in enumerate(video):
        if idx in compare_frames:
            frames.append(frame[...,::-1])

plt.figure(figsize=(8,8), facecolor=(1, 1, 1))

num_frames = len(compare_frames)
for i in range(num_frames):
    plt.subplot(num_frames, 2, 1 + 2*i)
    plt.imshow(frames1[i])
    plt.axis('off')
    if i == 0:
        plt.title(f'YOLOv5s\nframe {compare_frames[i]}')
    else:
        plt.title(f'frame {compare_frames[i]}')

    plt.subplot(num_frames, 2, 2 + 2*i)
    plt.imshow(frames2[i])
    plt.axis('off')
    if i==0:
        plt.title(f'YOLOv5x\nframe {compare_frames[i]}')
    else:
        plt.title(f'frame {compare_frames[i]}')

plt.tight_layout()
plt.savefig(save_name, dpi=200)
plt.show()
```
